## Finding Outliers 

```python
# Calculate IQR for 'Departure Delay'
Q1 = df['Departure Delay'].quantile(0.25)
Q3 = df['Departure Delay'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_departure = df[(df['Departure Delay'] < lower_bound) | (df['Departure Delay'] > upper_bound)]
print(f"Number of outliers in Departure Delay: {len(outliers_departure)}")

# Repeat for 'Arrival Delay'
Q1 = df['Arrival Delay'].quantile(0.25)
Q3 = df['Arrival Delay'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_arrival = df[(df['Arrival Delay'] < lower_bound) | (df['Arrival Delay'] > upper_bound)]
print(f"Number of outliers in Arrival Delay: {len(outliers_arrival)}")
```

## Graphs

```python
gender_counts = df['Gender'].value_counts()

# Create a pie chart
plt.figure(figsize=(5, 5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['green', 'orange'])
plt.title('Distribution of Passengers by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

```python
travel_counts = df['Type of Travel'].value_counts()

# Create a bar plot (histogram) for Type of Travel
plt.figure(figsize=(9, 5))
travel_counts.plot(kind='bar', color=['#66c2a5', '#fc8d62'])
plt.title('Distribution of Type of Travel')
plt.xlabel('Type of Travel')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
# Define a threshold for considering a flight as "canceled" based on departure delay
import matplotlib.pyplot as plt

cancellation_threshold = 180  # 180 minutes (3 hours)

# Determine if a flight is canceled
df['Canceled'] = (df['Departure Delay'] > cancellation_threshold) | (df['Arrival Delay'].isna())

# Count the number of canceled vs non-canceled flights
cancellation_counts = df['Canceled'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 6))
cancellation_counts.plot(kind='bar', color=['salmon', 'skyblue'], edgecolor='black')
plt.title('Flight Cancellations')
plt.xlabel('Flight Status')
plt.ylabel('Number of Flights')
plt.xticks(ticks=[0, 1], labels=['Non-Canceled', 'Canceled'], rotation=0)
plt.show()
```

```python
# Creating  histograms for Departure Delay and Arrival Delay

import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))

# Departure Delay histogram
plt.subplot(1, 2, 1)
plt.hist(df['Departure Delay'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Departure Delays')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Frequency')

# Arrival Delay histogram
plt.subplot(1, 2, 2)
plt.hist(df['Arrival Delay'].dropna(), bins=30, color='salmon', edgecolor='black')
plt.title('Histogram of Arrival Delays')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create a scatter plot for Departure Delay vs Arrival Delay
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Departure Delay', y='Arrival Delay', data=df, alpha=0.6)
plt.title('Scatter Plot of Departure Delay vs Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.axhline(0, color='red', linestyle='--', linewidth=1)  # Horizontal line at y=0
plt.axvline(0, color='red', linestyle='--', linewidth=1)  # Vertical line at x=0
plt.show()
```

```python
# Select relevant numerical columns for correlation
numeric_cols = ['Departure Delay', 'Arrival Delay', 'Flight Distance', 'In-flight Service', 
                'Food and Drink', 'Seat Comfort', 'Cleanliness']

# Calculate the correlation matrix
correlation_matrix = df[numeric_cols].corr()

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Heatmap of Correlation between Numerical Features')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('airline_passenger_satisfaction.csv')

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create a box plot for In-flight Service ratings based on Satisfaction Level
plt.figure(figsize=(10, 5))
sns.boxplot(x='Satisfaction', y='In-flight Service', data=data)
plt.title('Box Plot of In-flight Service Ratings by Satisfaction Level')
plt.xlabel('Satisfaction Level')
plt.ylabel('In-flight Service Ratings')
plt.xticks(rotation=45)
plt.show()
```
